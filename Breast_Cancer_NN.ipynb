```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
a=load_breast_cancer()
```

```python
df=pd.DataFrame(a.data,columns=a.feature_names)
```

```python
df['Target']=a.target
```

```python
y=df['Target']
x=df.drop("Target",axis=1)
```

```python
print(x)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
print(Xtrain.shape,Xtest.shape,Ytrain.shape,Ytest.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
Xtrain=scaler.fit_transform(Xtrain)
Xtest=scaler.fit_transform(Xtest)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
# this is used whenever we run the tensorflow it will randomly initialize the values so to avoid the change in accuracy whenever we train the model multiple times
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Sequential
```

```python
model=Sequential([
    #in the xtrain dataset we have 30 features so we have give 30 in the input_shape which tells the model takes the 30 feature
    Flatten(input_shape=(30,)),
    #Here after 30 we are giving no value because these model will accept the one dimensional data of 30 features
    Dense(20,activation="relu"),
    Dense(2,activation="softmax")
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",metrics=['accuracy'],optimizer="adam")
```

```python
histroy=model.fit(Xtrain,Ytrain,epochs=10,validation_split=0.1)
```

```python
plt.plot(histroy.history['accuracy'])
plt.plot(histroy.history['val_accuracy'])
plt.legend(['Actual Accuracy','Validation Accuracy'],loc="lower right")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy")
plt.show()
```

```python
plt.plot(histroy.history['loss'])
plt.plot(histroy.history['val_loss'])
plt.legend(['Actual loss','Validation loss'],loc="upper right")
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.title("loss")
plt.show()
```

```python
model.evaluate(Xtest,Ytest)
```

```python
y_pred=model.predict(Xtest)
```
